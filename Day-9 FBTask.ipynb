```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('fb.csv')
```

```python
#1 Count how many number of posts are paid and unpaid?

paid_count = df[df['Paid'] == True].shape[0]
unpaid_count = df[df['Paid'] == False].shape[0]
print("Number of paid posts:", paid_count)
print("Number of unpaid posts:", unpaid_count)
```

```python
#2 Display only Page type, Month, likes, comments and shares

df.iloc[:,[1,3,15,16,17]]
```

```python
#2 Display only Page type, Month, likes, comments and shares

df1 = df[['Type', 'Post Month', 'like', 'comment', 'share']]
df1
```

```python
#3 Find average number of likes to the post in September month

september_posts = df[df['Post Month'] == 9]
average_likes_september = september_posts['like'].mean()
print("Average number of likes to the post in September:", average_likes_september)
```

```python
#4 Print the information about the posts which are links.

link_posts = df[df['Type'] == 'Link']
print("Information about posts that are links:")
print(link_posts)
```

```python
#5 How many total shares are present for the video posts?

video_posts = df[df['Type'] == 'Video']
total_video_shares = video_posts['share'].sum()
print("Total shares present for video posts:", total_video_shares)
```

```python
#6 In which hour of the day maximum posts are posted?

hourly_post_counts = df.groupby('Post Hour').size()
max_hour = hourly_post_counts.idxmax()
max_posts = hourly_post_counts[max_hour]
print(f"The hour with the maximum number of posts is {max_hour} with {max_posts} posts.")
```

```python
#7 In which day of October, maximum number of interactions are found?

october_data = df[df['Post Month'] == 10]
daily_interactions = october_data.groupby(october_data['Post Month'])['Total Interactions'].sum()
max_day = daily_interactions.idxmax()
max_interactions = daily_interactions[max_day]
print(f"The day with the maximum number of interactions in October is {max_day} with {max_interactions} interactions.")
```

```python
#8 Print the information of the posts where likes are above 200 and shares are above 30.

filtered_posts = df[(df['like'] > 200) & (df['share'] > 30)]
print("Posts with likes above 200 and shares above 30:")
print(filtered_posts)
```

```python
#9 Which kind of posts are posted maximum number of times in November month?

november_data = df[df['Post Month'] == 11]
post_counts = november_data['Type'].value_counts()
max_post_type = post_counts.idxmax()
max_post_count = post_counts[max_post_type]
print(f"The type of post posted the maximum number of times in November is '{max_post_type}' with {max_post_count} occurrences.")
```

```python
#10 Find average and median of interactions for the paid (1) posts

paid_posts = df[df['Type'] == 1]
average_interactions = paid_posts['Total Interactions'].mean()
median_interactions = paid_posts['Total Interactions'].median()
print("Average interactions for paid posts:", average_interactions)
print("Median interactions for paid posts:", median_interactions)
```

```python
#11 Plot monthwise count of the posts.

import matplotlib.pyplot as plt
monthwise_count = df.groupby('Post Month').size()
plt.figure(figsize=(10, 6))
monthwise_count.plot(kind='bar', color='skyblue')
plt.title('Month-wise Count of Posts')
plt.xlabel('Month')
plt.ylabel('Count of Posts')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#12 Find the average of all reactions (like, comment and shares) of type ‘status’

status_entries = df[df['Type'] == 'Status']
average_reactions = status_entries[['like','comment','share']].mean()
print("Average of all reactions for status type:\n")
print(average_reactions)
```

```python
#13 Plot the monthwise total count of shares for photo posts.

photo_posts = df[(df['Type'] == 'Photo')]
monthwise_share_count = photo_posts.groupby('Post Month')['share'].sum()
plt.figure(figsize=(10, 6))
monthwise_share_count.plot(kind='bar', color='lightgreen')
plt.title('Month-wise Total Count of Shares for Photo Posts')
plt.xlabel('Month')
plt.ylabel('Total Count of Shares')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#14 Print the information of posts where no comments are posted. Which kind of posts are having highest no comments?

no_comments_posts = df[df['comment'] == 0]
print("Posts with no comments:")
print(no_comments_posts)
highest_no_comments_post_type = no_comments_posts['Type'].value_counts().idxmax()
highest_no_comments_count = no_comments_posts['post_type'].value_counts().max()
print("\nKind of posts with the highest number of no comments:")
print("Post Type:", highest_no_comments_post_type)
print("Number of No Comments:", highest_no_comments_count)
```

```python
#15  Sort all the posts in descending order of total interactions group by type during August – December

filtered_df = df[(df['Post Month'] >= 8) & (df['Post Month'] <= 12)]
grouped_df = filtered_df.groupby('Type')['Total Interactions'].sum()
sorted_df = grouped_df.sort_values(ascending=False)
print("Sorted posts in descending order of total interactions group by type during August – December:")
print(sorted_df)
```
