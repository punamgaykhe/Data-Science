# Subplot

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)#height,width,place

plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.plot(x,y,marker = '<',color='blue',mfc='r',ls='--',ms = 7);
plt.grid()

plt.subplot(2,2,3)
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.plot(x,z,marker = 'H',color='cyan',mfc='r',ls='--',ms = 11);
plt.grid()
```

# Bar Graph

```python
students=[56,65,46,51,61]

courses=['java','Python','R','Dart','Kotlin']
```

```python
plt.title('Bar Graph')
plt.xlabel('courses')
plt.ylabel('No. of students')
plt.bar(courses,students , color= 'pink',edgecolor='black',width=0.6,linewidth=3,alpha=0.7);
#alpha is used to color darkness
```

# Scatter Plot

```python
plt.title('Scatter Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.scatter(x,y,marker='*')
plt.grid()
```

# Pie Chart

```python
plt.title('Pie Chart')
plt.pie(students,labels=courses);
```

```python
plt.title('Pie Chart')
plt.pie(students,labels=courses,autopct='%.3f%%',shadow=True);
```

```python
plt.title('Pie Chart')
plt.pie(students,labels=courses,autopct='%.2f%%',shadow=True, radius=0.8,labeldistance=1.2,
        colors=['red','yellow','green','pink','black'],startangle=30);
```

```python
plt.title('Pie Chart')
plt.pie(students,labels=courses,autopct='%.2f%%',shadow=True, radius=0.8,labeldistance=1.2,
        colors=['red','yellow','green','pink','black'],startangle=30,
        explode=[0,0,0.2,0,0.3]);
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.plot(x,y,marker = '*',color='blue',mfc='r',ls='--',ms = 7);


plt.subplot(2,2,2)
plt.title('Bar Graph')
plt.xlabel('courses')
plt.ylabel('No. of students')
plt.bar(courses,students , color= 'pink',edgecolor='black',width=0.6,linewidth=3,alpha=0.7);

plt.subplot(2,2,3)
plt.title('Pie Chart')
plt.pie(students,labels=courses,autopct='%.2f%%',shadow=True, radius=1,labeldistance=1.1,
        colors=['red','yellow','green','pink','black'],startangle=30);

plt.subplot(2,2,4)
plt.title('Scatter Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.scatter(x,y,marker='*')
plt.grid()

plt.savefig('My Plot')
```

# Box Plot

```python
marks=[67,63,70,71,72,56,51,21,95]
```

```python
plt.boxplot(marks);
```

```python
q3=np.percentile(marks,75)
q1=np.percentile(marks,25)
```

```python
q3,q1
```

```python
iqr=q3-q1
iqr # interquartile range
```

```python
#lower Bound
lb=q1-1.5*iqr

#upper bound
up=q3+1.5*iqr
```

```python
lb,up
```

# Seaborn

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,hue='time')
plt.show()
```

```python
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
sns.histplot(tips['tip'],kde=True)
```

```python
plt.title('Outliers in tip')
plt.xlabel('Day')
plt.ylabel('tip')
sns.boxenplot(x='day',y='tip',data=tips);
plt.show()
```

```python
sns.pairplot(tips, hue='smoker');
plt.show()
```
